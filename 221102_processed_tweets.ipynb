```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
import nagisa
import re
```

```python
# Read the CSV file
df=pd.read_csv('210809-210905-raw-tweets.csv')
df.head()
```

```python
# Round1: Remove URL
df['Tweet']=df['Tweet'].str.replace('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ')
df.head()
```

```python
# Round2: Remove emoji
df['Tweet']=df['Tweet'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
df.head()
```

```python
# Round3: Remove spaces
df['Tweet'] = df['Tweet'].replace(' ', '', regex=True)
```

```python
# Round4: Remove newlines:\n
df['Tweet_processed'] = df['Tweet'].replace('\n','', regex=True)
df.head()
```
